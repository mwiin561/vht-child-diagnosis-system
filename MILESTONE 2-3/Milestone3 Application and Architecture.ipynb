### **System Architecture Explanation**

The system uses a hybrid cognitive AI design consisting of:

---

#### **1. NLP Understanding Engine**
- Accepts Luganda or English text.
- Extracts symptoms, checks negation, identifies age and illness duration.

---

#### **2. Machine Learning Classifier (Random Forest)**
- Accepts structured symptoms.
- Predicts probability scores for:
  - Pneumonia  
  - Malaria  
  - Diarrhea  

---

#### **3. Knowledge Graph Reasoning**
- Implements IMCI clinical rules.
- Weighted symptom-to-disease connections.
- Detects danger signs (convulsions, chest indrawing, etc.).
- Outputs symbolic reasoning scores.

---

#### **4. Hybrid Reasoning Layer**
Combines:
- **70% ML probability**
- **30% Knowledge Graph score**

This improves accuracy, interpretability, and clinical safety.

---

#### **5. Explanation Layer**
Produces:
- Detected symptoms
- Danger signs
- Classifier probabilities
- Knowledge Graph rule activations
- Final risk classification

---

#### **6. Application Layer (Streamlit UI)**
Allows VHTs to:
- Enter free text  
- View final diagnosis  
- Read explanation of reasoning  
- See risk level and next steps  

This architecture combines rule-based and machine learning intelligence into one coherent cognitive system.

```python
# =======================================================
# NLP SYMPTOM EXTRACTION FUNCTIONS
# =======================================================
# These functions convert free text (Luganda/English)
# into structured symptoms + metadata (age, duration).
# This replicates the Milestone 2 NLP engine.

import re

# Common negation words in Luganda and English
negation_words = {"no", "not", "without", "tet", "si", "siri", "sita", "siko"}

def clean_text(text):
    """Normalize text by removing symbols, lowering case, and trimming space."""
    text = text.lower()
    text = re.sub(r"[^a-z0-9\u0100-\uffff\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def is_negated(tokens, idx, window=3):
    """Returns True if the keyword is negated within ±3 words."""
    start = max(0, idx - window)
    end = min(len(tokens), idx + window + 1)
    return any(tokens[i] in negation_words for i in range(start, end))

def extract_symptoms(text):
    """
    Extract symptoms and clinical metadata (age, duration)
    from the user's free text description.
    """
    cleaned = clean_text(text)
    tokens = cleaned.split()

    # Initialize all symptoms as False
    symptoms = {sym: False for sym in symptom_map.keys()}

    # Keyword-based symptom matching
    for kw, symptom in keyword_lookup.items():
        if kw in cleaned:
            idx = cleaned.find(kw)
            token_index = len(cleaned[:idx].split())
            if not is_negated(tokens, token_index):
                symptoms[symptom] = True

    # Extract metadata
    age = 2
    duration = 1

    if m := re.search(r"(\d+)\s*year", cleaned):
        age = int(m.group(1))

    if m := re.search(r"(\d+)\s*(day|days|d)", cleaned):
        duration = int(m.group(1))

    metadata = {"age_years": age, "duration_days": duration}

    return symptoms, metadata
```

```python
# =======================================================
# KNOWLEDGE GRAPH + HYBRID REASONING MODULE
# =======================================================
# This reproduces the symbolic reasoning engine from Milestone 2
# and combines it with ML probabilities for a final cognitive output.

import networkx as nx
import pandas as pd

# Create Knowledge Graph
KG = nx.DiGraph()

diseases = ["pneumonia", "malaria", "diarrhea"]

danger_signs = [
    "convulsions", "chest_indrawing", "unable_to_feed",
    "vomiting_everything", "lethargic"
]

# Weighted IMCI rules
KG.add_edge("fast_breathing", "pneumonia", weight=0.9)
KG.add_edge("cough", "pneumonia", weight=0.7)
KG.add_edge("fever", "pneumonia", weight=0.4)

KG.add_edge("fever", "malaria", weight=0.8)
KG.add_edge("vomiting", "malaria", weight=0.6)
KG.add_edge("weakness", "malaria", weight=0.5)

KG.add_edge("diarrhea", "diarrhea", weight=0.9)
KG.add_edge("vomiting", "diarrhea", weight=0.5)
KG.add_edge("fever", "diarrhea", weight=0.3)


def kg_reasoning(symptoms):
    """
    Compute symbolic scores and identify danger signs
    using IMCI clinical rules.
    """
    scores = {d: 0 for d in diseases}
    explanations = []
    danger_hits = []

    # Add up weights for symptoms present
    for sym, present in symptoms.items():
        if present and sym in KG.nodes:
            for _, disease, data in KG.out_edges(sym, data=True):
                scores[disease] += data["weight"]
                explanations.append(f"{sym} → {disease} (weight {data['weight']})")

    # Identify danger signs
    for ds in danger_signs:
        if symptoms.get(ds, False):
            danger_hits.append(ds)

    # Determine risk level
    risk = (
        "high" if danger_hits
        else "moderate" if max(scores.values()) >= 0.7
        else "low"
    )

    return scores, danger_hits, risk, explanations


def hybrid_reasoning(symptoms, metadata):
    """
    Combine ML classifier probabilities and KG symbolic scores
    into a hybrid final prediction.
    """

    # Build feature vector in correct order
    fv = [
        int(symptoms["fever"]),
        int(symptoms["cough"]),
        int(symptoms["fast_breathing"]),
        int(symptoms["diarrhea"]),
        int(symptoms["vomiting"]),
        int(symptoms["weakness"]),
        int(symptoms["poor_feeding"]),
        int(symptoms["convulsions"]),
        metadata["age_years"],
        metadata["duration_days"]
    ]

    df = pd.DataFrame([fv])

    # ML Prediction
    probs = rf_model.predict_proba(df)[0]
    clf_pred = rf_model.predict(df)[0]
    clf_dict = {label: p for label, p in zip(rf_model.classes_, probs)}

    # KG Reasoning
    kg_scores, danger, risk, kg_rules = kg_reasoning(symptoms)

    # Hybrid Fusion
    final_scores = {d: 0.7*clf_dict[d] + 0.3*kg_scores[d] for d in diseases}
    final_pred = max(final_scores, key=final_scores.get)

    return final_pred, risk, clf_dict, kg_scores, kg_rules, danger
```

## **2. Streamlit Application (User Interface Layer)**

To complete the transformation of the cognitive agents developed in Milestone 2 into a usable system,  
a **Streamlit Application** was built. This application provides a clean, simple interface through which a VHT  
can input the child's symptoms (in Luganda or English) and receive a full diagnostic reasoning report.

### ⚡ Purpose of the Application Layer
The Streamlit app serves as the **front-end interface** for the cognitive system.  
It allows users to:

- Enter symptom descriptions in natural language  
- Trigger the NLP → ML → KG → Hybrid reasoning pipeline  
- View the final disease prediction  
- View risk level (low, moderate, high)  
- View full explanation (ML probabilities, KG rules fired, symptoms detected)  
- Identify any detected danger signs  

This layer enables **practical deployment**, making the cognitive system usable in the field by VHTs.

### 🧩 How It Works Internally
When a user clicks **Analyze**, the app:

1. Passes the text to the NLP module  
2. Extracts symptoms + metadata  
3. Runs ML classifier for probability scores  
4. Applies Knowledge Graph reasoning  
5. Combines both via the Hybrid reasoner  
6. Displays the final diagnosis + explanation  

This integrates **all Milestone 2 components** into a single operational system.

```python
%%writefile "/content/drive/MyDrive/Colab Notebooks/MILESTONE 2-3/app.py"

# ======================================================
# STREAMLIT APPLICATION — FINAL FULL VERSION (NO PLACEHOLDERS)
# VHT Childhood Disease Diagnosis Assistant
# ======================================================

import streamlit as st
import json
import joblib
import os
import numpy as np
import pandas as pd
import networkx as nx
import re

# ======================================================
# LOAD ARTIFACTS AND MODEL (ABSOLUTE PATHS)
# ======================================================

BASE_DIR = "/content/drive/MyDrive/Colab Notebooks/MILESTONE 2-3/Model Notebooks/"
ARTIFACT_DIR = BASE_DIR + "artifacts/"
MODEL_DIR = BASE_DIR + "model_artifacts/"

# Debug info (only visible inside Streamlit UI)
st.write("Loading artifacts from:", ARTIFACT_DIR)
st.write("Loading model from:", MODEL_DIR)

with open(ARTIFACT_DIR + "symptom_map.json", "r", encoding="utf-8") as f:
    symptom_map = json.load(f)

with open(ARTIFACT_DIR + "keyword_lookup.json", "r", encoding="utf-8") as f:
    keyword_lookup = json.load(f)

rf_model = joblib.load(MODEL_DIR + "child_disease_random_forest.pkl")

# ======================================================
# NLP FUNCTIONS
# ======================================================

negation_words = {"no", "not", "without", "tet", "si", "siri", "sita", "siko"}

def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z0-9\u0100-\uffff\s]", " ", text)
    return re.sub(r"\s+", " ", text).strip()

def is_negated(tokens, idx, window=3):
    start = max(0, idx - window)
    end = min(len(tokens), idx + window + 1)
    return any(tokens[i] in negation_words for i in range(start, end))

def extract_symptoms(text):
    cleaned = clean_text(text)
    tokens = cleaned.split()

    detected = {sym: False for sym in symptom_map.keys()}

    for kw, symptom in keyword_lookup.items():
        if kw in cleaned:
            idx = cleaned.find(kw)
            token_index = len(cleaned[:idx].split())
            if not is_negated(tokens, token_index):
                detected[symptom] = True

    age = 2
    duration = 1

    if m := re.search(r"(\d+)\s*year", cleaned):
        age = int(m.group(1))

    if m := re.search(r"(\d+)\s*(day|days|d)", cleaned):
        duration = int(m.group(1))

    return detected, {"age_years": age, "duration_days": duration}

# ======================================================
# KNOWLEDGE GRAPH RULES
# ======================================================

KG = nx.DiGraph()

KG.add_edge("fast_breathing", "pneumonia", weight=0.9)
KG.add_edge("cough", "pneumonia", weight=0.7)
KG.add_edge("fever", "pneumonia", weight=0.4)

KG.add_edge("fever", "malaria", weight=0.8)
KG.add_edge("vomiting", "malaria", weight=0.6)
KG.add_edge("weakness", "malaria", weight=0.5)

KG.add_edge("diarrhea", "diarrhea", weight=0.9)
KG.add_edge("vomiting", "diarrhea", weight=0.5)
KG.add_edge("fever", "diarrhea", weight=0.3)

diseases = ["pneumonia", "malaria", "diarrhea"]

danger_signs = [
    "convulsions", "chest_indrawing", "unable_to_feed",
    "vomiting_everything", "lethargic"
]

def kg_reasoning(symptoms):
    scores = {d: 0 for d in diseases}
    explanations = []
    danger = []

    for s, present in symptoms.items():
        if present and s in KG.nodes:
            for _, disease, data in KG.out_edges(s, data=True):
                scores[disease] += data["weight"]
                explanations.append(f"{s} → {disease} (weight {data['weight']})")

    for ds in danger_signs:
        if symptoms.get(ds, False):
            danger.append(ds)

    risk = "high" if danger else "moderate" if max(scores.values()) >= 0.7 else "low"
    return scores, danger, risk, explanations

# ======================================================
# HYBRID REASONING ENGINE (ML + KG)
# ======================================================

def hybrid_reasoning(symptoms, metadata):

    fv = [
        int(symptoms["fever"]),
        int(symptoms["cough"]),
        int(symptoms["fast_breathing"]),
        int(symptoms["diarrhea"]),
        int(symptoms["vomiting"]),
        int(symptoms["weakness"]),
        int(symptoms["poor_feeding"]),
        int(symptoms["convulsions"]),
        metadata["age_years"],
        metadata["duration_days"]
    ]

    df = pd.DataFrame([fv], columns=[
        "fever", "cough", "fast_breathing", "diarrhea", "vomiting",
        "weakness", "poor_feeding", "convulsions",
        "age_years", "duration_days"
    ])

    probs = rf_model.predict_proba(df)[0]
    clf_pred = rf_model.predict(df)[0]
    clf_probs = {label: float(p) for label, p in zip(rf_model.classes_, probs)}

    kg_scores, danger, risk, kg_rules = kg_reasoning(symptoms)

    final_scores = {d: 0.7*clf_probs[d] + 0.3*kg_scores[d] for d in diseases}
    final_pred = max(final_scores, key=final_scores.get)

    return final_pred, risk, clf_probs, kg_scores, kg_rules, danger

# ======================================================
# STREAMLIT USER INTERFACE (FINAL)
# ======================================================

st.title("🩺 VHT Childhood Disease Diagnosis Assistant")
st.write("Enter symptoms in Luganda or English, or select a test case from the sidebar.")

# Sidebar Predefined Test Cases
st.sidebar.header("🧪 Test Cases")

test_cases = {
    "Select a test case": "",
    "Pneumonia (Fast Breathing)": "Omwana afuuya mangu era akyawa okukosora.",
    "Malaria (Fever + Weakness)": "Omwana alina omusujja era alina obunafu.",
    "Diarrhea Case": "Omwana alina endwadde y'ekidugavu era ali kusgala.",
    "Danger Signs (Convulsions)": "Omwana alina obutonya obungi era tayinza kulya.",
    "English Case (Malaria-like)": "The child has had a high fever for 3 days and is vomiting.",
    "English Case (Pneumonia-like)": "The child is breathing very fast and coughing."
}

selected_case = st.sidebar.selectbox("Choose Test Case:", list(test_cases.keys()))

input_default = test_cases[selected_case]

# Main Text Area
input_text = st.text_area("Symptoms:", value=input_default, height=160)

# PROCESS
if st.button("Analyze"):

    if not input_text.strip():
        st.warning("⚠️ Please enter symptoms or choose a test case.")
        st.stop()

    symptoms, metadata = extract_symptoms(input_text)
    final_pred, risk, clf_probs, kg_scores, kg_rules, danger = hybrid_reasoning(symptoms, metadata)

    st.subheader("Final Diagnosis")
    st.success(final_pred.upper())

    st.subheader("Risk Level")
    st.warning(risk.upper())

    st.subheader("Detected Symptoms")
    st.json(symptoms)

    st.subheader("ML Probabilities")
    st.json(clf_probs)

    st.subheader("Knowledge Graph Rules Fired")
    for rule in kg_rules:
        st.write("•", rule)

    if danger:
        st.error(f"🚨 DANGER SIGNS DETECTED: {', '.join(danger)}")
```
